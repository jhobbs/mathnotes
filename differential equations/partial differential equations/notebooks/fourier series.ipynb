```python
# Import necessary libraries
from sage.symbolic.integration.integral import definite_integral
from sage.symbolic.ring import var

# Define the function and the variable
x = var('x', domain='real')
f = -4*x**2  # Example function, you can replace it with any periodic function

# Define the period
T = 2 * pi  # Period of the function
L = T / 2

# Compute Fourier coefficients
a0 = (1 / L) * definite_integral(f, x, -L, L)
an = lambda n: (1 / L) * definite_integral(f * cos(n * pi * x / L), x, -L, L)
bn = lambda n: (1 / L) * definite_integral(f * sin(n * pi * x / L), x, -L, L)

m = var('m', domain='integer')

show((an(m).simplify() * cos(m * pi * x / L) + bn(m) * sin(m * pi * x / L)).simplify())
# Construct the Fourier series
N = 10  # Number of terms in the Fourier series
fourier_series = a0 / 2 + sum(an(n) * cos(n * pi * x / L) + bn(n) * sin(n * pi * x / L) for n in range(1, N + 1))

show(fourier_series)

# Plot the original function and Fourier series
p1 = plot(f, (x, -L, L), color='red', legend_label='Original Function')
p2 = plot(fourier_series, (x, -L, L), color='blue', legend_label='Fourier Series (N=10)')

# Show the plot
(p1 + p2).show()
```

```python
# Define variables
x, n = var('x n')
L = pi  # Half the period, you can adjust this according to your function's period

# Define a symbolic function f(x)
f = e**(3*x)

# Symbolic Fourier coefficients
a0 = (1 / L) * integral(f, (x, -L, L))
an = (1 / L) * integral(f * cos(n * pi * x / L), (x, -L, L))
bn = (1 / L) * integral(f * sin(n * pi * x / L), (x, -L, L))

# Create the LaTeX expression for the Fourier series
fourier_series_latex = LatexExpr(r"a_0/2 + \sum_{n=1}^\infty \left( a_n \cos\left(\frac{n \pi x}{L}\right) + b_n \sin\left(\frac{n \pi x}{L}\right) \right)")

# Substitute symbols with their definitions
fourier_series_latex = fourier_series_latex.subs(a0=a0, an=an, bn=bn)

# Display the formatted Fourier series
show(fourier_series_latex)
```
